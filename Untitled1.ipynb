```python
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt

# Univariate Regression
X = np.array([[1],[2],[3],[4],[5]])
y = np.array([2,4,6,8,10])

model = LinearRegression()
model.fit(X,y)

pred = model.predict(X)

print("Prediction:", model.predict([[6]]))

# Plot
plt.scatter(X,y)
plt.plot(X,pred)
plt.title("Univariate Regression")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([1,2,3,4,5])
y = np.array([2,4,5,4,5])

m = ((x-x.mean())*(y-y.mean())).sum()/((x-x.mean())**2).sum()
c = y.mean() - m*x.mean()

y_pred = m*x + c

print("Slope:",m)
print("Intercept:",c)

# Plot
plt.scatter(x,y)
plt.plot(x,y_pred)
plt.title("Least Square Regression")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt

data = load_iris()

X = data.data[:, :2]
y = data.target

model = LogisticRegression(max_iter=200)
model.fit(X,y)

print("Accuracy:", model.score(X,y))

# Plot
plt.scatter(X[:,0], X[:,1], c=y)
plt.title("Logistic Regression")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

```python
from sklearn.linear_model import Perceptron
import numpy as np
import matplotlib.pyplot as plt

X = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([0,0,0,1])

model = Perceptron()
model.fit(X,y)

print("Prediction:", model.predict([[1,1]]))

# Plot
plt.scatter(X[:,0], X[:,1], c=y)
plt.title("Single Layer Perceptron")
plt.xlabel("Input 1")
plt.ylabel("Input 2")
plt.show()
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt

data = load_iris()

X = data.data[:, :2]
y = data.target

model = MLPClassifier(hidden_layer_sizes=(5,), max_iter=1000)

model.fit(X,y)

print("Accuracy:", model.score(X,y))

# Plot
plt.scatter(X[:,0], X[:,1], c=y)
plt.title("MLP Classifier")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

```python
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt

X = np.array([[1,2],[1,4],[1,0],
              [10,2],[10,4],[10,0]])

model = KMeans(n_clusters=2)

model.fit(X)

print("Labels:", model.labels_)

# Plot
plt.scatter(X[:,0], X[:,1], c=model.labels_)
plt.title("K-Means Clustering")
plt.show()
```

```python
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt

data = load_iris()

X = data.data
y = data.target

pca = PCA(n_components=2)

X_pca = pca.fit_transform(X)

print("Reduced Shape:", X_pca.shape)

# Plot
plt.scatter(X_pca[:,0], X_pca[:,1], c=y)
plt.title("PCA")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.show()
```

```python
from sklearn.datasets import load_iris
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt
import pandas as pd


iris = load_iris()

df = pd.DataFrame(iris.data, columns=iris.feature_names)


X1 = df[['sepal length (cm)']]
y1 = df['petal length (cm)']

model1 = LinearRegression()
model1.fit(X1, y1)

pred1 = model1.predict(X1)

print("Univariate R2 Score:", r2_score(y1, pred1))


plt.scatter(X1, y1)
plt.plot(X1, pred1)
plt.title("Univariate Regression")
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")
plt.show()


X2 = df[['sepal length (cm)', 'sepal width (cm)']]
y2 = df['petal length (cm)']

model2 = LinearRegression()
model2.fit(X2, y2)

pred2 = model2.predict(X2)

print("Bivariate R2 Score:", r2_score(y2, pred2))


plt.scatter(df['sepal length (cm)'], y2)
plt.title("Bivariate Regression")
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")
plt.show()


X3 = df.iloc[:, :3]
y3 = df['petal width (cm)']

model3 = LinearRegression()
model3.fit(X3, y3)

pred3 = model3.predict(X3)

print("Multivariate R2 Score:", r2_score(y3, pred3))


plt.plot(y3.values, label='Actual')
plt.plot(pred3, label='Predicted')
plt.title("Multivariate Regression")
plt.legend()
plt.show()
```
